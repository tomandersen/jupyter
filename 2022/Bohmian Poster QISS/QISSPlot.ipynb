# Plot quantum witness operator vs drop time

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math

mu = 0
std = 1
x = np.linspace(start=0, stop=4*math.pi, num=500)

plt.rcParams['figure.figsize'] = [40, 18]
plt.rcParams['font.size'] = '40'

y = 2*(np.sin(x/2))**2 
qg, = plt.plot(x, y, 'c--', linewidth=5, label='quantum gravity')

y = 0.5*(np.sin(x/2))**2 + (np.sin(x/4))**2
btg, = plt.plot(x, y, '-.', linewidth=5, label='Bohmian trajectory gravity')

y = 0*x
semi, = plt.plot(x, y, linewidth=5, label='semi classical gravity')

plt.xlabel('Drop Time', fontsize=40)
plt.ylabel('Entanglement Witness Value', fontsize=40)


plt.legend(handles=[qg, btg, semi], loc='upper right', fontsize=35 )

plt.savefig('QISSWitness.png')


plt.show()


print(0.5*(np.sin(math.pi/2))**2 + (np.sin(math.pi/4))**2)
```
