# From position to momentum representation

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

hbar=1
```

# Position wavefunction localized at one point in space with zero uncertainty. 

```python
L=1.0; dx=0.01; Nx=int( L/dx + 1 )

x=np.linspace(0.0, L, Nx)

def delta(x,x0):
    if x==x0:
        delta=1
    else:
        delta=0
    return delta

psi=np.zeros(Nx)

for ix in range(Nx):
    psi[ix]=delta(x[ix],0.50)
        
plt.plot(x,np.abs(psi)**2)

plt.xlabel("$x$", fontsize=14)
plt.ylabel("$|\psi(x)|^2$", fontsize=14)
plt.savefig('psi_x1.png') 
plt.show()
```

```python
maxp=100.0; dp=0.01; Np=int( (2*maxp)/dp + 1)

p=np.linspace(-maxp, maxp, Np)
phi=np.zeros(Np,dtype='complex')

for ip in range(Np):
    phi[ip]=0.0
    for ix in range(Nx):
        phi[ip]=phi[ip] + dx * psi[ix] * (1/np.sqrt(2*np.pi*hbar)) * cmath.exp(1j*x[ix]*p[ip]/hbar)
        
plt.plot(p,np.abs(phi)**2)

plt.xlabel("$p$", fontsize=14)
plt.ylabel("$|\phi(p)|^2$", fontsize=14)
plt.savefig('phi_p1.png') 
plt.show()
```

When a particle is localized at one point in space, corresponding to zero uncertainty, its momentum distribution is uniform amounting to infinite uncertainty.

# The momentum wavefunction shows quasi-localization behaviour when summed over several position eigenstates

```python
phi=np.zeros(Np,dtype='complex')

for ip in range(Np):
    phi[ip]=0.0
    for ix in range(Nx):
        phi[ip]=phi[ip] + dx * psi[ix] * (1/np.sqrt(2*np.pi*hbar)) * cmath.exp(1j*x[ix]*p[ip]/hbar)
        
plt.plot(p,np.abs(phi)**2)

plt.xlabel("$p$", fontsize=14)
plt.ylabel("$|\phi(p)|^2$", fontsize=14)
plt.savefig('phi_p2.png') 
plt.show()
```
