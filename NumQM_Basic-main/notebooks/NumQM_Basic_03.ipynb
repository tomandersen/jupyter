```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
    
L=1.0; dx=0.01; N=int( 2*L/dx + 1 )

x=np.linspace(0, L, N)
psi_1=np.sqrt(2/L)*np.sin(1*x*np.pi/L)
psi_4=np.sqrt(2/L)*np.sin(4*x*np.pi/L)
psi_10=np.sqrt(2/L)*np.sin(10*x*np.pi/L)

plt.plot(x,psi_1**2)
plt.plot(x,psi_4**2)
plt.plot(x,psi_10**2)
plt.xlabel("$x$", fontsize=14)
plt.ylabel("$|\psi(x)|^2$", fontsize=14)
plt.legend(['n=1','n=4','n=10'])
plt.savefig('psi_x.png') 
plt.show()
```

```python
hbar=1
def delta_x_p(x0,p):
    delta_x_p=(1/np.sqrt(2*np.pi*hbar)) * cmath.exp(-1j*x0*p/hbar)
    return delta_x_p
    
A=100.0; dp=0.05; Np=int( (2*A)/dx + 1)

p=np.linspace(-A, A, Np)
phi_p1=np.zeros(Np,dtype='complex')
phi_p4=np.zeros(Np,dtype='complex')
phi_p10=np.zeros(Np,dtype='complex')

for i in range(Np):
    phi_p1[i]=0
    phi_p4[i]=0
    phi_p10[i]=0
    for k in range(N):
        phi_p1[i]=phi_p1[i] + dx * psi_1[k] * delta_x_p(x[k],p[i])
        phi_p4[i]=phi_p4[i] + dx * psi_4[k] * delta_x_p(x[k],p[i])
        phi_p10[i]=phi_p10[i] + dx * psi_10[k] * delta_x_p(x[k],p[i])

plt.plot(p,np.abs(phi_p1)**2)
plt.plot(p,np.abs(phi_p4)**2)
plt.plot(p,np.abs(phi_p10)**2)

plt.xlabel("$p$", fontsize=14)
plt.ylabel("$|\phi(p)|^2$", fontsize=14)
plt.legend(['n=1','n=4','n=10'])
plt.savefig('phi_p.png') 
plt.show()
```
